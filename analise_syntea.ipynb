```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV das condições
conditions = pd.read_csv('synthea_sample_data_csv_apr2020/csv/conditions.csv')

# Exibir as primeiras linhas para entender o formato
print(conditions.head())

# Agrupar por nome da condição e contar a frequência
condition_counts = conditions['DESCRIPTION'].value_counts()

# Exibir as 10 condições mais comuns
print(condition_counts.head(10))

# Plotar um gráfico das condições mais comuns
condition_counts.head(10).plot(kind='bar', figsize=(10, 5), title='Condições Médicas Mais Comuns')
plt.xlabel('Condição')
plt.ylabel('Frequência')
plt.show()
```

```python
# Carregar dados de vacinação
immunizations = pd.read_csv('synthea_sample_data_csv_apr2020/csv/immunizations.csv')

# Contar frequência de vacinas
vaccine_counts = immunizations['DESCRIPTION'].value_counts()

# Exibir as vacinas mais comuns
print(vaccine_counts.head(10))

# Plotar as vacinas mais comuns
vaccine_counts.head(10).plot(kind='bar', figsize=(10, 5), title='Vacinas Mais Comuns')
plt.xlabel('Vacina')
plt.ylabel('Frequência')
plt.show()
```
